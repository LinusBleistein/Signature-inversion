# Inverting fake signatures generated by WGAN with Gradient Penalty

```python
%matplotlib inline

import torch
import numpy as np
import signatory
import math

import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from insertion import Insertion, get_A_matrix, solve_optimization_problem,invert_signature
from tools import get_signature_as_tensor
```

## Import WGAN data

```python
fake_sig = pd.read_csv('data/Fake signature/fake_sig.csv', header=None)
```

## Inverting and plotting of fake inverted signatures

```python
from mpl_toolkits import mplot3d

for i in np.random.choice(np.arange(0,1000),size=6):
    
    test_sig = torch.tensor(np.array(fake_sig.iloc[i,1:]))
    test_sig = test_sig.view(1,120)
    reconstructed_path=invert_signature(test_sig,3,0,3)
    
    fig = plt.subplots(figsize=(5,5))
    ax = plt.axes(projection='3d')
    ax.view_init(0, 10)
    ax.plot3D(reconstructed_path[0,:], reconstructed_path[1,:], reconstructed_path[2,:],marker='o',linewidth=2, linestyle='--')
    ax.set_title('Path ' + str(i))
    
    fig = plt.subplots(3,figsize=(5,5))
    ax = plt.axes(projection='3d')
    ax.view_init(40, 20)
    ax.plot3D(reconstructed_path[0,:], reconstructed_path[1,:], reconstructed_path[2,:],marker='o',linewidth=2,linestyle='--')
    
    fig = plt.subplots(3,figsize=(5,5))
    ax = plt.axes(projection='3d')
    ax.view_init(60, 30)
    ax.plot3D(reconstructed_path[0,:], reconstructed_path[1,:], reconstructed_path[2,:],marker='o',linewidth=2,linestyle='--')

```
