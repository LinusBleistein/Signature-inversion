# Inverting fake signatures generated by WGAN with Gradient Penalty

```python
%matplotlib inline

import torch
import numpy as np
import signatory
import math

import pandas as pd

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from insertion import Insertion, get_A_matrix, solve_optimization_problem,invert_signature
from tools import get_signature_as_tensor
```

## Inverting generated signatures of order 4 for integer 0

The code bellow inverts fake signatures generated by a WGAN-GP. The network tries to mimic signatures of handdrawn 0's coming from the database MNIST. Generated signatures are taken at order 4.

```python
fake_sig_int_0_order_4 = pd.read_csv('data/Fake integer signature/fake_sig_integer_0_order_4.csv', header=None)
```

```python
j = -1

f,ax = plt.subplots(3,3,figsize=(25,25))

f.suptitle('Inverted signatures of order 5, for integer 1', fontsize='xx-large')

for i in np.random.choice(np.arange(0,1000),size=9):
    
    j= j + 1
    
    test_sig = torch.tensor(np.array(fake_sig_int_0_order_4.iloc[i,1:]))
    test_sig = test_sig.view(1,-1)
    reconstructed_path=invert_signature(test_sig,3,0,2)
    ax[j%3,j//3].plot(reconstructed_path[:,0],reconstructed_path[:,1],marker='o')

plt.savefig('Inverted 0 order 4')
plt.show()
```
